## Invoking the Tool
Invoking the tool is the same across all EMU tools. The tool determines what to plot based on the user-specified EMU run directory name. For example, 'emu_samp...' corresponds to the Sampling Tool, while 'emu_fgrd...' corresponds to the Forward Gradient Tool.

There are two method to use the tool:
- **Method 1: Menu-driven Input**
  
`runpy.run_path('/efs_ecco/ECCO/EMU/emu_userinterface_dir/python/emu_plot.py');`

or `globals_dict = ept.emu_plot()` by using the ept module.

This method is menu-driven and does not specify any arguments. This method is most suitable for users who are new to the EMU tool. The menu-driven method guides users through entering parameters and creating plots. 

- **Method 2: Argument-based Input**

`globals_dict = ept.emu_plot(run_name="/PATH/emu_fgrd_7_15_743_5_-1.00E-01")`

This method uses the ept moudle and skips the step of creating plots in Method 1. Instead, it focuses on returning data for users to perform more advanced processing later. The arguments provide more flexible ways for users to extract data.

For further details, including the available arguments, you can use 

`help(ept.emu_plot)`.

```python
## To make a higher-quality plot
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# take care of the unit: 
def convert_VT_to_OHT(vt, rho=1024, cp=3990):
    # vt: the origional output variables, e.g., "v_prime * T_bar" and "v_bar * T_prime"
    #     where "var_prime" is object variable selected in EMU 
    #     and "var_bar" is the climatology-based weight specificed in the mask file
    return vt*cp*rho/1e15 # unit: PW

atrb = return_vars_dict['variable']
atrb = convert_VT_to_OHT(atrb)
residual = atrb[0,:]-np.sum(atrb[1:,:],axis=0)
var_names = ['Total','Wind','Heat Flux','Freshwater Flux','Initial Condition','Residual']
data2plot = [atrb[0,:],atrb[1,:],atrb[2,:],atrb[3,:],atrb[6,:],residual]
time = pd.date_range(start='1/1/1992', end='1/1/2018',freq='ME')
colors = ['k', 'forestgreen', 'crimson', 'royalblue', 'salmon', 'gold']

fig, ax = plt.subplots(1,1,figsize=(7,4))
ax.axhline(y=0,lw=0.5,c='k')
ax.axvspan(time[215],time[217],fc='silver',alpha=0.5)
for i in range(len(var_names)):
    plt.plot(time, data2plot[i], label=var_names[i], c=colors[i])
plt.title('30N Atlantic Upper 1000m MHT Anomalies due to V Anomalies')
plt.legend(ncol=2)
plt.xlabel('time')
plt.ylabel('[PW]')
plt.xlim(time[0],time[-1])
plt.ylim(-0.04,0.027)
plt.tight_layout()
plt.savefig('/efs_ecco/yfan/plots/Attribution_MHT30N_upper1000m_V.png',dpi=1000)
plt.show()
```
